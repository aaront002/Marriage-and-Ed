# Preparing Data for Group Project

```python
import pandas as pd
from urllib.request import urlopen
mared = pd.read_pickle(urlopen('https://github.com/aaront002/Marriage-and-Ed/raw/master/mared.pkl'),compression=None)
```

Filtering data to only retrieve fertility data for group project 

```python
mared.where(mared["SeriesName"]=="Fertility rate, total (births per woman)", inplace = True) 
```

```python
mared
```

```python
mared.tail()
```

## Continent Data

```python
link1='https://en.wikipedia.org/wiki/List_of_sovereign_states_and_dependent_territories_by_continent_(data_file)'
cont=pd.read_html(link1,header=0,flavor='bs4',attrs={'class': 'wikitable sortable'})[0]
```

```python
cont
```

```python
cont.drop(columns=['a-2','#'])
```

```python
cont.drop(columns=['a-2','#'], inplace= True)
```

```python
cont.head()
```

Changing header names

```python
OldToNew={cont.columns[0]:'ContinentCode',
          cont.columns[1]:'CountryCode',
          cont.columns[2]:'CountryName'}
```

```python
cont.rename(columns=OldToNew,inplace=True)
```

```python
cont.head()
```

## Merging the Data

```python
cont.drop(columns=['CountryName'], inplace= True)
```

```python
merge1=cont.merge(mared,on="CountryCode")
```

```python
merge1
```

```python
contmared=merge1.iloc[:222,:].copy()
```

```python
contmared
```

```python
import numpy as np
contmared.ContinentCode=[str(x) for x in contmared.ContinentCode]
```

```python
contmared.ContinentCode.value_counts()
```

NA for North America is being registered as "Not Available"

```python
#change North America being registered from 'nan' to 'NA'
replacements={'nan': 'NA'}
contmared.ContinentCode.replace(replacements, inplace=True)
```

```python
#double-checking if the changes worked
contmared.ContinentCode.value_counts()
```

```python
contmared.dtypes
```

```python
contmared
```

Dropping 1990 and 2010 data in order to more easily merge with team's data and conduct analysis for group project

```python
contmared.drop(columns=['SeriesName','1990','2010'], inplace=True)
```

```python
contmared.head()
```

```python
#renaming year 2000 to Fertility Rate since group data will all be from 2010
OldToNew2={contmared.columns[3]:'FertilityRate'}
```

```python
contmared.rename(columns=OldToNew2,inplace=True)
```

```python
#There are symbols that represent missing values, but they are not recognized so. We need to check what values in those numeric columns are not numbers:
badmaredSymbols=[] # list for bad symbols

NumericColNames=mared.iloc[:,3:].columns # save names of columns with numeric data

for columnName in NumericColNames:# visit every column name
    for cell in mared[columnName]:# visit every cell for that column
        try:
            float(cell) # try this
        except: # if not possible:            
            if cell not in badmaredSymbols:# if cell is not in the list                
                badmaredSymbols.append(cell)# add it to the list

# you get:
badmaredSymbols
```

```python
import numpy as np

mared.replace(to_replace=badmaredSymbols,value=np.nan,inplace=True)
```
